```python
import pandas as pd
from ydata_profiling import ProfileReport
```

### Load the data

```python
df = pd.read_csv("Popular_Baby_Names.csv")
```

```python
#to analyze descriptive statistics
df.describe()
```

```python
#Generate the profile report
profile = ProfileReport(df, title='Pandas Profiling Report')
```

```python
#Display the report
profile.to_notebook_iframe()
```

```python
# to save the profiling results
profile.to_file("fikayomi.html")
```
